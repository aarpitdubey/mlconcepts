# Gradient Descent

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data using a dummy function
X = np.arange(10)
Y = (X-5)**2 + 3

print(X, Y)
```

```python
# Visualization of the curve

plt.style.use('seaborn-v0_8-whitegrid')
plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Gradient Descent with negative slope
# initialize x with any random value
x = 0
# learning rate
lr = 0.1

plt.plot(X, Y)

#loop
for i in range(50):
    # gradient
    gradient = 2 * (x-5)
    # Udation of x
    x = x - lr * gradient
    y = (x-5)**2 + 3
    plt.scatter(x, y)
    print(x)
plt.show()
```

```python
# Gradient Descent with positive slope
# initialize x with any random value
x = 9
# learning rate
lr = 0.1

plt.plot(X, Y)

#loop
for i in range(50):
    # gradient
    gradient = (2 * (x-5))
    # Udation of x
    x = x - lr * gradient
    y = (x-5)**2 + 3
    plt.scatter(x, y)
    print(x)
plt.show()
```
